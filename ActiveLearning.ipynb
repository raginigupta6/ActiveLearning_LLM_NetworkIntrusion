```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

def preprocess_data(df):
    df_copy = df.copy()
    X = df_copy.drop('Label', axis=1)
    y = df_copy['Label']

    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    return X.values, y.values

def train_and_evaluate(X_train, X_test, y_train, y_test, hidden_layers=(100, 100), max_iter=300):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    mlp = MLPClassifier(hidden_layer_sizes=hidden_layers, max_iter=max_iter, random_state=42)
    mlp.fit(X_train, y_train)

    y_pred = mlp.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    return accuracy, f1, conf_matrix, class_report, scaler, mlp

def calculate_fpr_fnr(conf_matrix):
    FP = conf_matrix.sum(axis=0) - np.diag(conf_matrix)
    FN = conf_matrix.sum(axis=1) - np.diag(conf_matrix)
    TP = np.diag(conf_matrix)
    TN = conf_matrix.sum() - (FP + FN + TP)

    FPR = np.divide(FP, FP + TN, out=np.zeros_like(FP, dtype=float), where=(FP + TN) != 0)
    FNR = np.divide(FN, FN + TP, out=np.zeros_like(FN, dtype=float), where=(FN + TP) != 0)

    return FPR, FNR

def calculate_overall_fpr_fnr(FPR, FNR, conf_matrix):
    total_samples = conf_matrix.sum()
    class_counts = conf_matrix.sum(axis=1)

    overall_fpr = np.sum(FPR * class_counts) / total_samples
    overall_fnr = np.sum(FNR * class_counts) / total_samples

    return overall_fpr, overall_fnr

def print_evaluation_results(name, accuracy, f1, conf_matrix, class_report):
    print(f"Accuracy on {name}: {accuracy:.4f}")
    print(f"F1 Score on {name}: {f1:.4f}")
    print(f"Confusion Matrix on {name}:")
    print(conf_matrix)
    print(f"Classification Report on {name}:")
    print(class_report)

    FPR, FNR = calculate_fpr_fnr(conf_matrix)
    overall_fpr, overall_fnr = calculate_overall_fpr_fnr(FPR, FNR, conf_matrix)

    print(f"False Positive Rate (FPR) on {name}: {FPR}")
    print(f"False Negative Rate (FNR) on {name}: {FNR}")
    print(f"Overall False Positive Rate (FPR) on {name}: {overall_fpr:.4f}")
    print(f"Overall False Negative Rate (FNR) on {name}: {overall_fnr:.4f}")

# Assuming df2017 and df2018 are your dataframes
X_2017, y_2017 = preprocess_data(df2017)
X_2018, y_2018 = preprocess_data(df2018)

# Split the 2017 data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)

# Train and evaluate on 2017 data
accuracy_2017, f1_2017, conf_matrix_2017, class_report_2017, scaler, mlp = train_and_evaluate(X_train, X_test, y_train, y_test)

# Print the evaluation results for the 2017 data
print_evaluation_results("df2017", accuracy_2017, f1_2017, conf_matrix_2017, class_report_2017)

# Standardize the 2018 data using the same scaler fitted on the 2017 training data
X_2018 = scaler.transform(X_2018)

# Make predictions and evaluate on the 2018 data
y_pred_2018 = mlp.predict(X_2018)
accuracy_2018 = accuracy_score(y_2018, y_pred_2018)
f1_2018 = f1_score(y_2018, y_pred_2018, average='weighted')
conf_matrix_2018 = confusion_matrix(y_2018, y_pred_2018)
class_report_2018 = classification_report(y_2018, y_pred_2018)

# Print the evaluation results for the 2018 data
print_evaluation_results("df2018", accuracy_2018, f1_2018, conf_matrix_2018, class_report_2018)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras import layers, models
import tensorflow.keras.backend as K
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def preprocess_data(df):
    df_copy = df.copy()
    X = df_copy.drop('Label', axis=1)
    y = df_copy['Label']

    # Convert features to numeric, handling errors and missing values
    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    # Convert string labels to numbers
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    return X.values, y

def create_contrastive_autoencoder(input_dim):
    input_layer = layers.Input(shape=(input_dim,))
    encoder = layers.Dense(128, activation='relu')(input_layer)
    encoder = layers.Dense(64, activation='relu')(encoder)
    encoder = layers.Dense(32, activation='relu')(encoder)

    decoder = layers.Dense(64, activation='relu')(encoder)
    decoder = layers.Dense(128, activation='relu')(decoder)
    decoder = layers.Dense(input_dim, activation='sigmoid')(decoder)

    autoencoder = models.Model(inputs=input_layer, outputs=decoder)
    encoder_model = models.Model(inputs=input_layer, outputs=encoder)

    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder, encoder_model

def contrastive_loss(y_true, y_pred):
    margin = 1
    square_pred = K.square(y_pred)
    margin_square = K.square(K.maximum(margin - y_pred, 0))
    y_true = K.cast(y_true, y_pred.dtype)  # Ensure y_true is the same type as y_pred
    return K.mean(y_true * square_pred + (1 - y_true) * margin_square)


def create_pairs(X, y):
    pairs = []
    labels = []
    num_classes = len(np.unique(y))
    digit_indices = [np.where(y == i)[0] for i in range(num_classes)]

    for d in range(num_classes):
        if len(digit_indices[d]) < 2:
            print(f"Class {d} has less than 2 samples, skipping pair creation for this class.")
            continue
        for i in range(len(digit_indices[d]) - 1):
            z1, z2 = digit_indices[d][i], digit_indices[d][i + 1]
            pairs.append([X[z1], X[z2]])
            labels.append(1)
            inc = np.random.randint(1, num_classes)
            dn = (d + inc) % num_classes
            if len(digit_indices[dn]) == 0:
                continue
            z1, z2 = digit_indices[d][i], digit_indices[dn][i % len(digit_indices[dn])]
            pairs.append([X[z1], X[z2]])
            labels.append(0)

    print(f"Generated {len(pairs)} pairs.")
    return np.array(pairs), np.array(labels)

def train_contrastive_autoencoder(autoencoder, X_train, y_train, epochs=15, batch_size=256):
    pairs, labels = create_pairs(X_train, y_train)

    if len(pairs) == 0 or len(labels) == 0:
        raise ValueError("Generated pairs or labels are empty. Check the data and pair generation process.")

    input_a = layers.Input(shape=(X_train.shape[1],))
    input_b = layers.Input(shape=(X_train.shape[1],))

    processed_a = autoencoder(input_a)
    processed_b = autoencoder(input_b)

    distance = layers.Lambda(euclidean_distance)([processed_a, processed_b])

    model = models.Model([input_a, input_b], distance)
    model.compile(optimizer='adam', loss=contrastive_loss)

    validation_split = 0.1
    split_index = int(len(pairs) * (1 - validation_split))

    train_pairs = pairs[:split_index]
    val_pairs = pairs[split_index:]
    train_labels = labels[:split_index]
    val_labels = labels[split_index:]

    if len(train_pairs) == 0 or len(val_pairs) == 0:
        raise ValueError("Insufficient data for training or validation after splitting.")

    history = model.fit([train_pairs[:, 0], train_pairs[:, 1]], train_labels,
                        validation_data=([val_pairs[:, 0], val_pairs[:, 1]], val_labels),
                        batch_size=batch_size, epochs=epochs, verbose=1)

    return history

def euclidean_distance(vects):
    x, y = vects
    return K.sqrt(K.sum(K.square(x - y), axis=1, keepdims=True))

def plot_loss(history):
    plt.plot(history.history['loss'], label='train_loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.legend()
    plt.show()

def detect_drift_samples(encoder, X_train, y_train, X_test, TMAD=3.5):
    encoded_train = encoder.predict(X_train)
    encoded_test = encoder.predict(X_test)

    N = len(np.unique(y_train))
    centroids = []
    MADs = []

    for i in range(N):
        class_samples = encoded_train[y_train == i]
        centroid = np.mean(class_samples, axis=0)
        distances = np.linalg.norm(class_samples - centroid, axis=1)
        MAD = np.median(np.abs(distances - np.median(distances))) * 1.4826
        centroids.append(centroid)
        MADs.append(MAD)

    drift_indices = []

    for idx, sample in enumerate(encoded_test):
        distances = [np.linalg.norm(sample - centroid) for centroid in centroids]
        A = np.min([(distances[i] - np.median(distances[i])) / MADs[i] for i in range(N)])
        if A > TMAD:
            drift_indices.append(idx)

    return drift_indices

# Preprocess data
X_2017, y_2017 = preprocess_data(df2017)
X_2018, y_2018 = preprocess_data(df2018)

# Normalize the data
scaler = StandardScaler()
X_2017 = scaler.fit_transform(X_2017)
X_2018 = scaler.transform(X_2018)

# Split data into training and testing sets
X_train_2017, X_test_2017, y_train_2017, y_test_2017 = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)
X_train_2018, X_test_2018, y_train_2018, y_test_2018 = train_test_split(X_2018, y_2018, test_size=0.2, random_state=42)

# Train autoencoder on 2017 data
input_dim = X_train_2017.shape[1]
autoencoder, encoder = create_contrastive_autoencoder(input_dim)
history = train_contrastive_autoencoder(autoencoder, X_train_2017, y_train_2017, epochs=15, batch_size=256)

# Plot loss
plot_loss(history)

# Detect drift samples in 2018 data
drift_indices_2018 = detect_drift_samples(encoder, X_train_2017, y_train_2017, X_train_2018)
print(f"Number of detected drift samples: {len(drift_indices_2018)}")

X_2018_train_sample = X_train_2018[drift_indices_2018]
y_2018_train_sample = y_train_2018[drift_indices_2018]

# Combine the 2017 training data with the detected drift samples from 2018
X_combined_train = np.vstack((X_train_2017, X_2018_train_sample))
y_combined_train = np.concatenate((y_train_2017, y_2018_train_sample))

# Train and evaluate on the combined dataset
accuracy_combined, f1_combined, conf_matrix_combined, class_report_combined, scaler_combined, mlp_combined = train_and_evaluate(X_combined_train, X_test_2018, y_combined_train, y_test_2018)

# Print the evaluation results for the 2018 test data
print_evaluation_results("df2018", accuracy_combined, f1_combined, conf_matrix_combined, class_report_combined)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

def preprocess_data(df):
    df_copy = df.copy()
    X = df_copy.drop('Label', axis=1)
    y = df_copy['Label']

    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    return X.values, y.values

def train_and_evaluate(X_train, X_test, y_train, y_test, hidden_layers=(100, 100), max_iter=300):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    mlp = MLPClassifier(hidden_layer_sizes=hidden_layers, max_iter=max_iter, random_state=42)
    mlp.fit(X_train, y_train)

    y_pred = mlp.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    return accuracy, f1, conf_matrix, class_report, scaler, mlp

def calculate_fpr_fnr(conf_matrix):
    FP = conf_matrix.sum(axis=0) - np.diag(conf_matrix)
    FN = conf_matrix.sum(axis=1) - np.diag(conf_matrix)
    TP = np.diag(conf_matrix)
    TN = conf_matrix.sum() - (FP + FN + TP)

    FPR = np.divide(FP, FP + TN, out=np.zeros_like(FP, dtype=float), where=(FP + TN) != 0)
    FNR = np.divide(FN, FN + TP, out=np.zeros_like(FN, dtype=float), where=(FN + TP) != 0)

    return FPR, FNR

def calculate_overall_fpr_fnr(FPR, FNR, conf_matrix):
    total_samples = conf_matrix.sum()
    class_counts = conf_matrix.sum(axis=1)

    overall_fpr = np.sum(FPR * class_counts) / total_samples
    overall_fnr = np.sum(FNR * class_counts) / total_samples

    return overall_fpr, overall_fnr

def print_evaluation_results(name, accuracy, f1, conf_matrix, class_report):
    print(f"Accuracy on {name}: {accuracy:.4f}")
    print(f"F1 Score on {name}: {f1:.4f}")
    print(f"Confusion Matrix on {name}:")
    print(conf_matrix)
    print(f"Classification Report on {name}:")
    print(class_report)

    FPR, FNR = calculate_fpr_fnr(conf_matrix)
    overall_fpr, overall_fnr = calculate_overall_fpr_fnr(FPR, FNR, conf_matrix)

    print(f"False Positive Rate (FPR) on {name}: {FPR}")
    print(f"False Negative Rate (FNR) on {name}: {FNR}")
    print(f"Overall False Positive Rate (FPR) on {name}: {overall_fpr:.4f}")
    print(f"Overall False Negative Rate (FNR) on {name}: {overall_fnr:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

def preprocess_data(df):
    df_copy = df.copy()
    X = df_copy.drop('Label', axis=1)
    y = df_copy['Label']

    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    return X.values, y.values

def train_and_evaluate(X_train, X_test, y_train, y_test, hidden_layers=(100, 100), max_iter=300):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    mlp = MLPClassifier(hidden_layer_sizes=hidden_layers, max_iter=max_iter, random_state=42)
    mlp.fit(X_train, y_train)

    y_pred = mlp.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    return accuracy, f1, conf_matrix, class_report, scaler, mlp

def calculate_fpr_fnr(conf_matrix):
    FP = conf_matrix.sum(axis=0) - np.diag(conf_matrix)
    FN = conf_matrix.sum(axis=1) - np.diag(conf_matrix)
    TP = np.diag(conf_matrix)
    TN = conf_matrix.sum() - (FP + FN + TP)

    FPR = np.divide(FP, FP + TN, out=np.zeros_like(FP, dtype=float), where=(FP + TN) != 0)
    FNR = np.divide(FN, FN + TP, out=np.zeros_like(FN, dtype=float), where=(FN + TP) != 0)

    return FPR, FNR

def calculate_overall_fpr_fnr(FPR, FNR, conf_matrix):
    total_samples = conf_matrix.sum()
    class_counts = conf_matrix.sum(axis=1)

    overall_fpr = np.sum(FPR * class_counts) / total_samples
    overall_fnr = np.sum(FNR * class_counts) / total_samples

    return overall_fpr, overall_fnr

def print_evaluation_results(name, accuracy, f1, conf_matrix, class_report):
    print(f"Accuracy on {name}: {accuracy:.4f}")
    print(f"F1 Score on {name}: {f1:.4f}")
    print(f"Confusion Matrix on {name}:")
    print(conf_matrix)
    print(f"Classification Report on {name}:")
    print(class_report)

    FPR, FNR = calculate_fpr_fnr(conf_matrix)
    overall_fpr, overall_fnr = calculate_overall_fpr_fnr(FPR, FNR, conf_matrix)

    print(f"False Positive Rate (FPR) on {name}: {FPR}")
    print(f"False Negative Rate (FNR) on {name}: {FNR}")
    print(f"Overall False Positive Rate (FPR) on {name}: {overall_fpr:.4f}")
    print(f"Overall False Negative Rate (FNR) on {name}: {overall_fnr:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from scipy.stats import entropy
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

# Preprocess data
X_2017, y_2017 = preprocess_data(df2017)
X_2018, y_2018 = preprocess_data(df2018)

# Split the data into training and testing sets
X_train_2017, X_test_2017, y_train_2017, y_test_2017 = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)
X_train_2018, X_test_2018, y_train_2018, y_test_2018 = train_test_split(X_2018, y_2018, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler().fit(X_train_2017)
X_train_2017_scaled = scaler.transform(X_train_2017)
X_train_2018_scaled = scaler.transform(X_train_2018)
X_test_2018_scaled = scaler.transform(X_test_2018)

# Train initial MLP model on 2017 data
mlp_initial = MLPClassifier(hidden_layer_sizes=(100, 100), random_state=42, max_iter=500)
mlp_initial.fit(X_train_2017_scaled, y_train_2017)

def extract_features(model, data):
    activation = data
    for i in range(len(model.coefs_) - 1):
        activation = activation @ model.coefs_[i] + model.intercepts_[i]
        activation = np.maximum(activation, 0)  # ReLU activation function
    return activation

# Extract feature embeddings for the 2018 data
X_train_2018_embeddings = extract_features(mlp_initial, X_train_2018_scaled)

# Predict probabilities for the 2018 training data
probabilities_2018 = mlp_initial.predict_proba(X_train_2018_scaled)

# Calculate entropy for each sample
uncertainties = entropy(probabilities_2018, axis=1)

# Number of clusters (equal to the number of samples we want to select)
num_clusters = int(0.01 * len(X_train_2018))

# Perform weighted K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit K-Means on the uncertainty-weighted embeddings
weighted_embeddings = X_train_2018_embeddings * uncertainties[:, np.newaxis]
kmeans.fit(weighted_embeddings)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Find the nearest neighbors to the cluster centroids
distances = euclidean_distances(weighted_embeddings, centroids)
nearest_neighbors = np.argmin(distances, axis=0)

# Select the most informative samples
X_2018_train_sample = X_train_2018[nearest_neighbors]
y_2018_train_sample = y_train_2018[nearest_neighbors]

# Combine the 2017 training data with the selected samples from 2018
X_combined_train = np.vstack((X_train_2017, X_2018_train_sample))
y_combined_train = np.concatenate((y_train_2017, y_2018_train_sample))

# Train and evaluate on the combined dataset
accuracy_combined, f1_combined, conf_matrix_combined, class_report_combined, scaler_combined, mlp_combined = train_and_evaluate(X_combined_train, X_test_2018_scaled, y_combined_train, y_test_2018)

# Print the evaluation results for the 2018 data
print_evaluation_results("2018 Dataset", accuracy_combined, f1_combined, conf_matrix_combined, class_report_combined)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

def preprocess_data(df):
    df_copy = df.copy()
    X = df_copy.drop('Label', axis=1)
    y = df_copy['Label']

    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    return X.values, y.values

def train_and_evaluate(X_train, X_test, y_train, y_test, hidden_layers=(100, 100), max_iter=300):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    mlp = MLPClassifier(hidden_layer_sizes=hidden_layers, max_iter=max_iter, random_state=42)
    mlp.fit(X_train, y_train)

    y_pred = mlp.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    return accuracy, f1, conf_matrix, class_report, scaler, mlp

def calculate_fpr_fnr(conf_matrix):
    FP = conf_matrix.sum(axis=0) - np.diag(conf_matrix)
    FN = conf_matrix.sum(axis=1) - np.diag(conf_matrix)
    TP = np.diag(conf_matrix)
    TN = conf_matrix.sum() - (FP + FN + TP)

    FPR = np.divide(FP, FP + TN, out=np.zeros_like(FP, dtype=float), where=(FP + TN) != 0)
    FNR = np.divide(FN, FN + TP, out=np.zeros_like(FN, dtype=float), where=(FN + TP) != 0)

    return FPR, FNR

def calculate_overall_fpr_fnr(FPR, FNR, conf_matrix):
    total_samples = conf_matrix.sum()
    class_counts = conf_matrix.sum(axis=1)

    overall_fpr = np.sum(FPR * class_counts) / total_samples
    overall_fnr = np.sum(FNR * class_counts) / total_samples

    return overall_fpr, overall_fnr

def print_evaluation_results(name, accuracy, f1, conf_matrix, class_report):
    print(f"Accuracy on {name}: {accuracy:.4f}")
    print(f"F1 Score on {name}: {f1:.4f}")
    print(f"Confusion Matrix on {name}:")
    print(conf_matrix)
    print(f"Classification Report on {name}:")
    print(class_report)

    FPR, FNR = calculate_fpr_fnr(conf_matrix)
    overall_fpr, overall_fnr = calculate_overall_fpr_fnr(FPR, FNR, conf_matrix)

    print(f"False Positive Rate (FPR) on {name}: {FPR}")
    print(f"False Negative Rate (FNR) on {name}: {FNR}")
    print(f"Overall False Positive Rate (FPR) on {name}: {overall_fpr:.4f}")
    print(f"Overall False Negative Rate (FNR) on {name}: {overall_fnr:.4f}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers, models
import tensorflow.keras.backend as K
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

# Preprocessing function
def preprocess_data(df):
    df_copy = df.copy()
    X = df_copy.drop('Label', axis=1)
    y = df_copy['Label']

    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    return X.values, y

# Random selection for active learning
def select_random_samples(X, y, percentage=0.01):
    num_samples = int(len(X) * percentage)
    indices = np.random.choice(range(len(X)), num_samples, replace=False)
    return X[indices], y[indices], indices

# Creating positive and negative pairs
def create_pairs(X_2017, y_2017, X_2018, y_2018):
    pairs = []
    labels = []

    for x_2018, y_2018_label in zip(X_2018, y_2018):
        positive_indices = np.where(y_2017 == y_2018_label)[0]
        negative_indices = np.where(y_2017 != y_2018_label)[0]

        num_positive_pairs = min(10, len(positive_indices))
        num_negative_pairs = min(10, len(negative_indices))

        if num_positive_pairs > 0:
            positive_samples = np.random.choice(positive_indices, num_positive_pairs, replace=False)
            for idx in positive_samples:
                pairs.append([x_2018, X_2017[idx]])
                labels.append(1)

        if num_negative_pairs > 0:
            negative_samples = np.random.choice(negative_indices, num_negative_pairs, replace=False)
            for idx in negative_samples:
                pairs.append([x_2018, X_2017[idx]])
                labels.append(0)

    return np.array(pairs), np.array(labels)

# Creating embedding network
def create_embedding_network(input_dim, num_classes):
    input_layer = layers.Input(shape=(input_dim,))
    x = layers.Dense(100, activation='relu')(input_layer)
    x = layers.Dense(100, activation='relu')(x)
    embeddings = x
    classification_output = layers.Dense(num_classes, activation='softmax', name='classification_output')(embeddings)
    return models.Model(inputs=input_layer, outputs=[classification_output, embeddings])

# Euclidean distance for contrastive loss
def euclidean_distance(vects):
    x, y = vects
    return K.sqrt(K.sum(K.square(x - y), axis=1, keepdims=True))

# Contrastive loss function
def contrastive_loss(y_true, y_pred):
    margin = 1
    square_pred = K.square(y_pred)
    margin_square = K.square(K.maximum(margin - y_pred, 0))
    y_true = K.cast(y_true, y_pred.dtype)  # Ensure y_true is the same type as y_pred
    return K.mean(y_true * square_pred + (1 - y_true) * margin_square)

# Custom training loop to handle both losses
def train_model(model, X_train, y_train, pairs, labels, epochs=15, batch_size=256, contrastive_loss_weight=0.5):
    optimizer = tf.keras.optimizers.Adam()
    num_classes = len(np.unique(y_train))

    # Creating dataset for combined training
    X_a = pairs[:, 0]
    X_b = pairs[:, 1]
    y_combined_train = np.concatenate((y_train, np.zeros(len(pairs))))
    X_combined_train = np.concatenate((X_train, X_a))
    X_combined_b = np.concatenate((np.zeros_like(X_train), X_b))
    y_combined_labels = np.concatenate((np.zeros(len(y_train)), labels))

    for epoch in range(epochs):
        print(f'Epoch {epoch+1}/{epochs}')

        # Shuffle the combined training data
        indices = np.random.permutation(len(X_combined_train))
        X_combined_train_shuffled = X_combined_train[indices]
        X_combined_b_shuffled = X_combined_b[indices]
        y_combined_train_shuffled = y_combined_train[indices]
        y_combined_labels_shuffled = y_combined_labels[indices]

        with tf.GradientTape() as tape:
            # Forward pass
            logits, embeddings_a = model(X_combined_train_shuffled, training=True)
            embeddings_b = model(X_combined_b_shuffled, training=True)[1]

            # Classification loss
            classification_loss = tf.keras.losses.sparse_categorical_crossentropy(y_combined_train_shuffled, logits)

            # Contrastive loss
            distance = euclidean_distance([embeddings_a, embeddings_b])
            contrastive_loss_value = contrastive_loss(y_combined_labels_shuffled, distance)

            # Combined loss
            total_loss = classification_loss + contrastive_loss_weight * contrastive_loss_value

        gradients = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

        # Print loss information
        print(f'Combined Loss: {tf.reduce_mean(total_loss).numpy()}')

    return model

# Evaluate performance
def evaluate_performance(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred)
    return accuracy, f1, conf_matrix, class_report

X_2017, y_2017 = preprocess_data(df2017)
X_2018, y_2018 = preprocess_data(df2018)

# Normalize the data
scaler = StandardScaler()
X_2017 = scaler.fit_transform(X_2017)
X_2018 = scaler.transform(X_2018)

# Check for NaNs and Infs in the data
if np.any(np.isnan(X_2017)) or np.any(np.isinf(X_2017)) or np.any(np.isnan(X_2018)) or np.any(np.isinf(X_2018)):
    raise ValueError("Data contains NaNs or Infs")

# Split data into training and testing sets
X_train_2017, X_test_2017, y_train_2017, y_test_2017 = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)
X_train_2018, X_test_2018, y_train_2018, y_test_2018 = train_test_split(X_2018, y_2018, test_size=0.2, random_state=42)

# Select 1% random samples from 2018
X_2018_train_sample, y_2018_train_sample, _ = select_random_samples(X_train_2018, y_train_2018, percentage=0.01)

# Combine 2017 data with 1% 2018 data
X_combined_train = np.vstack((X_train_2017, X_2018_train_sample))
y_combined_train = np.concatenate((y_train_2017, y_2018_train_sample))

# Create pairs for contrastive learning
pairs, labels = create_pairs(X_train_2017, y_train_2017, X_2018_train_sample, y_2018_train_sample)

# Create and train model
input_dim = X_combined_train.shape[1]
num_classes = len(np.unique(y_combined_train))
model = create_embedding_network(input_dim, num_classes)

# Train the model with both losses
model = train_model(model, X_combined_train, y_combined_train, pairs, labels, epochs=15, batch_size=256, contrastive_loss_weight=0.5)

# Predict classes for the 2018 test data
logits_2018, _ = model.predict(X_test_2018)
y_pred_2018_classes = np.argmax(logits_2018, axis=1)

# Evaluate the performance on 2018 test data
accuracy_2018, f1_2018, conf_matrix_2018, class_report_2018 = evaluate_performance(y_test_2018, y_pred_2018_classes)

print(f"Accuracy on 2018 test data: {accuracy_2018}")
print(f"F1 Score on 2018 test data: {f1_2018}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import torch.nn.functional as F

class FeatureExtractor(nn.Module):
    def __init__(self, input_dim):
        super(FeatureExtractor, self).__init__()
        self.fc1 = nn.Linear(input_dim, 100)
        self.fc2 = nn.Linear(100, 100)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return x

class Classifier(nn.Module):
    def __init__(self, feature_dim, output_dim):
        super(Classifier, self).__init__()
        self.fc = nn.Linear(feature_dim, output_dim)

    def forward(self, x):
        x = self.fc(x)
        return x

class DomainDiscriminator(nn.Module):
    def __init__(self, feature_dim):
        super(DomainDiscriminator, self).__init__()
        self.fc1 = nn.Linear(feature_dim, 100)
        self.fc2 = nn.Linear(100, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

def train_dann(feature_extractor, classifier, domain_discriminator, source_loader, target_loader, epochs):
    criterion_class = nn.CrossEntropyLoss()
    criterion_domain = nn.BCELoss()

    optimizer_feature = optim.Adam(feature_extractor.parameters(), lr=0.001)
    optimizer_class = optim.Adam(classifier.parameters(), lr=0.001)
    optimizer_domain = optim.Adam(domain_discriminator.parameters(), lr=0.001)

    for epoch in range(epochs):
        feature_extractor.train()
        classifier.train()
        domain_discriminator.train()

        total_class_loss = 0.0
        total_domain_loss = 0.0

        for (source_data, source_labels), (target_data, _) in zip(source_loader, target_loader):
            source_data, source_labels = source_data.to(device), source_labels.to(device)
            target_data = target_data.to(device)

            # Train feature extractor and classifier
            optimizer_feature.zero_grad()
            optimizer_class.zero_grad()

            source_features = feature_extractor(source_data)
            class_outputs = classifier(source_features)
            class_loss = criterion_class(class_outputs, source_labels)

            class_loss.backward(retain_graph=True)
            optimizer_feature.step()
            optimizer_class.step()

            total_class_loss += class_loss.item()

            # Train domain discriminator
            optimizer_feature.zero_grad()
            optimizer_domain.zero_grad()

            target_features = feature_extractor(target_data)
            source_domain_labels = torch.ones(source_features.size(0), 1).to(device)
            target_domain_labels = torch.zeros(target_features.size(0), 1).to(device)

            domain_outputs_source = domain_discriminator(source_features.detach())
            domain_outputs_target = domain_discriminator(target_features.detach())

            domain_loss_source = criterion_domain(domain_outputs_source, source_domain_labels)
            domain_loss_target = criterion_domain(domain_outputs_target, target_domain_labels)

            domain_loss = (domain_loss_source + domain_loss_target) / 2
            domain_loss.backward()

            optimizer_feature.step()
            optimizer_domain.step()

            total_domain_loss += domain_loss.item()

        print(f'Epoch [{epoch+1}/{epochs}], Classification Loss: {total_class_loss/len(source_loader)}, Domain Loss: {total_domain_loss/len(source_loader)}')

# Example usage with CIC IDS 2017 and 2018 datasets
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Assuming X_train_2017, y_train_2017, X_train_2018 are preprocessed and available as tensors
source_dataset = TensorDataset(torch.tensor(X_train_2017, dtype=torch.float32), torch.tensor(y_train_2017, dtype=torch.long))
target_dataset = TensorDataset(torch.tensor(X_train_2018, dtype=torch.float32), torch.tensor(y_train_2018, dtype=torch.long))

source_loader = DataLoader(source_dataset, batch_size=32, shuffle=True)
target_loader = DataLoader(target_dataset, batch_size=32, shuffle=True)

input_dim = X_train_2017.shape[1]
output_dim = len(np.unique(y_train_2017))

feature_extractor = FeatureExtractor(input_dim).to(device)
classifier = Classifier(100, output_dim).to(device)
domain_discriminator = DomainDiscriminator(100).to(device)

train_dann(feature_extractor, classifier, domain_discriminator, source_loader, target_loader, epochs=50)
```

```python
def evaluate_dann(feature_extractor, classifier, domain_discriminator, source_loader, target_loader):
    feature_extractor.eval()
    classifier.eval()
    domain_discriminator.eval()

    source_correct = 0
    target_correct = 0
    total_source = 0
    total_target = 0

    y_true = []
    y_pred = []

    # Classification evaluation on the source domain
    with torch.no_grad():
        for source_data, source_labels in source_loader:
            source_data, source_labels = source_data.to(device), source_labels.to(device)
            source_features = feature_extractor(source_data)
            class_outputs = classifier(source_features)
            _, predicted = torch.max(class_outputs.data, 1)
            total_source += source_labels.size(0)
            source_correct += (predicted == source_labels).sum().item()
            y_true.extend(source_labels.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

    source_accuracy = source_correct / total_source
    print(f'Source Domain Accuracy: {source_accuracy * 100:.2f}%')

    # Print classification report and confusion matrix
    print("Classification Report for Source Domain:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix for Source Domain:")
    print(confusion_matrix(y_true, y_pred))

    y_true_domain = []
    y_pred_domain = []

    # Domain discrimination evaluation
    with torch.no_grad():
        for source_data, _ in source_loader:
            source_data = source_data.to(device)
            source_features = feature_extractor(source_data)
            domain_outputs_source = domain_discriminator(source_features)
            y_true_domain.extend(torch.ones(source_data.size(0)).cpu().numpy())
            y_pred_domain.extend(torch.round(domain_outputs_source).cpu().numpy())

        for target_data, _ in target_loader:
            target_data = target_data.to(device)
            target_features = feature_extractor(target_data)
            domain_outputs_target = domain_discriminator(target_features)
            y_true_domain.extend(torch.zeros(target_data.size(0)).cpu().numpy())
            y_pred_domain.extend(torch.round(domain_outputs_target).cpu().numpy())

    print("Classification Report for Domain Discrimination:")
    print(classification_report(y_true_domain, y_pred_domain))
    print("Confusion Matrix for Domain Discrimination:")
    print(confusion_matrix(y_true_domain, y_pred_domain))

# Assuming X_test_2017, y_test_2017, X_test_2018, and y_test_2018 are preprocessed and available as tensors
source_test_dataset = TensorDataset(torch.tensor(X_test_2017, dtype=torch.float32), torch.tensor(y_test_2017, dtype=torch.long))
target_test_dataset = TensorDataset(torch.tensor(X_test_2018, dtype=torch.float32), torch.tensor(y_test_2018, dtype=torch.long))

source_test_loader = DataLoader(source_test_dataset, batch_size=32, shuffle=False)
target_test_loader = DataLoader(target_test_dataset, batch_size=32, shuffle=False)

evaluate_dann(feature_extractor, classifier, domain_discriminator, source_test_loader, target_test_loader)
```
