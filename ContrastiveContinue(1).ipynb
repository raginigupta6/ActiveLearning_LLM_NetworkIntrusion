Preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

def preprocess_data(df):
    # Create a copy of the DataFrame for binary classification
    df_binary = df.copy()

    # Map labels to binary classes: 'benign' and 'malicious'
    df_binary['Label'] = df_binary['Label'].apply(lambda x: 'benign' if x == 'BENIGN' or x == 'Benign' else 'malicious')

    # Separate the features and the target
    X = df_binary.drop('Label', axis=1)
    y = df_binary['Label']

    # Convert categorical labels to binary
    y = y.map({'benign': 0, 'malicious': 1})

    # Convert all columns to numeric, coerce errors to NaN, and fill NaNs with the mean of the column
    X = X.apply(pd.to_numeric, errors='coerce')
    X.fillna(X.mean(), inplace=True)

    # Replace infinities with large finite numbers
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.mean(), inplace=True)

    return X, y
```

Encoder + MLP

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

def create_autoencoder(input_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(128, activation='relu')(input_layer)
    encoded = Dense(64, activation='relu')(encoded)
    encoded = Dense(32, activation='relu')(encoded)
    decoded = Dense(64, activation='relu')(encoded)
    decoded = Dense(128, activation='relu')(decoded)
    decoded = Dense(input_dim, activation='sigmoid')(decoded)
    autoencoder = Model(input_layer, decoded)
    encoder = Model(input_layer, encoded)
    return autoencoder, encoder

def extract_embeddings(model, data):
    return model.predict(data)

X_2017, y_2017 = preprocess_data(df2017)

# Split the 2017 data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)

# Standardize the feature data (important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create and train the autoencoder
input_dim = X_train.shape[1]
autoencoder, encoder = create_autoencoder(input_dim)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X_train, X_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Extract embeddings for training and testing data
X_train_encoded = extract_embeddings(encoder, X_train)
X_test_encoded = extract_embeddings(encoder, X_test)

# Initialize the MLP model
mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=300, random_state=42)

# Train the MLP model on the encoded training data
mlp.fit(X_train_encoded, y_train)

# Make predictions on the encoded test data
y_pred = mlp.predict(X_test_encoded)

# Evaluate the model on the 2017 test set
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Calculate FPR and FNR
tn, fp, fn, tp = conf_matrix.ravel()
fpr = fp / (fp + tn)
fnr = fn / (fn + tp)

# Print the evaluation results for the 2017 test set
print(f"Accuracy on df2017: {accuracy:.4f}")
print(f"F1 Score on df2017: {f1:.4f}")
print(f"FPR on df2017: {fpr:.4f}")
print(f"FNR on df2017: {fnr:.4f}")
print("Confusion Matrix on df2017:")
print(conf_matrix)
print("Classification Report on df2017:")
print(class_report)
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.callbacks import EarlyStopping

def contrastive_loss(y_true, y_pred, margin=1.0):
    y_true = tf.cast(y_true, tf.float32)
    square_pred = tf.square(y_pred)
    margin_square = tf.square(tf.maximum(margin - y_pred, 0))
    return tf.reduce_mean(y_true * square_pred + (1 - y_true) * margin_square)

def generate_contrastive_pairs(X, y):
    pairs = []
    labels = []
    n = len(y)

    for i in range(n):
        # Find all positive and negative samples
        positive_indices = np.where(y == y.iloc[i])[0]
        negative_indices = np.where(y != y.iloc[i])[0]

        # Remove self-index from positive indices
        positive_indices = positive_indices[positive_indices != i]

        # Randomly select one positive and one negative sample
        if len(positive_indices) > 0:
            pos_idx = np.random.choice(positive_indices)
            pairs.append([X.iloc[i].values, X.iloc[pos_idx].values])
            labels.append(1)

        if len(negative_indices) > 0:
            neg_idx = np.random.choice(negative_indices)
            pairs.append([X.iloc[i].values, X.iloc[neg_idx].values])
            labels.append(0)

    pairs = np.array(pairs)
    labels = np.array(labels)

    return [pairs[:, 0], pairs[:, 1]], labels

X_2017, y_2017 = preprocess_data(df2017)

# Split the 2017 data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)

# Standardize the feature data (important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert X_train and y_train to DataFrame and Series for proper indexing
X_train_df = pd.DataFrame(X_train, columns=X_2017.columns)
y_train_series = pd.Series(y_train)

# Generate contrastive pairs for training
[pairs_a, pairs_b], labels = generate_contrastive_pairs(X_train_df, y_train_series)

# Define the autoencoder
input_dim = X_train.shape[1]
autoencoder, encoder = create_autoencoder(input_dim)
autoencoder.compile(optimizer='adam', loss=MeanSquaredError())

# Define a custom training loop for contrastive learning
class ContrastiveAutoencoder(tf.keras.Model):
    def __init__(self, encoder, autoencoder, margin=1.0):
        super(ContrastiveAutoencoder, self).__init__()
        self.encoder = encoder
        self.autoencoder = autoencoder
        self.margin = margin

    def compile(self, optimizer, loss_fn):
        super(ContrastiveAutoencoder, self).compile()
        self.optimizer = optimizer
        self.loss_fn = loss_fn

    def call(self, inputs, training=False):
        input_a, input_b = inputs
        encoded_a = self.encoder(input_a)
        encoded_b = self.encoder(input_b)
        reconstructed_a = self.autoencoder(input_a)
        reconstructed_b = self.autoencoder(input_b)
        return encoded_a, encoded_b, reconstructed_a, reconstructed_b


    def train_step(self, data):
        input_a, input_b = data[0]
        y = data[1]

        with tf.GradientTape() as tape:
            # Forward pass
            encoded_a = self.encoder(input_a)
            encoded_b = self.encoder(input_b)
            reconstructed_a = self.autoencoder(input_a)
            reconstructed_b = self.autoencoder(input_b)

            # Compute reconstruction loss
            recon_loss = self.loss_fn(input_a, reconstructed_a) + self.loss_fn(input_b, reconstructed_b)

            # Compute contrastive loss
            dist = tf.reduce_sum(tf.square(encoded_a - encoded_b), axis=1)
            contrastive_loss_value = contrastive_loss(y, dist, self.margin)

            # Total loss
            total_loss = recon_loss + contrastive_loss_value

        # Compute gradients
        gradients = tape.gradient(total_loss, self.autoencoder.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.autoencoder.trainable_variables))

        return {"loss": total_loss, "recon_loss": recon_loss, "contrastive_loss": contrastive_loss_value}

# Initialize the contrastive autoencoder
contrastive_autoencoder = ContrastiveAutoencoder(encoder, autoencoder)
contrastive_autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss_fn=MeanSquaredError())

# Train the model
contrastive_autoencoder.fit(
    x=[pairs_a, pairs_b],
    y=labels,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3)]
)

# Extract embeddings for training and testing data
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# Initialize the MLP model
mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=300, random_state=42)

# Train the MLP model on the encoded training data
mlp.fit(X_train_encoded, y_train)

# Make predictions on the encoded test data
y_pred = mlp.predict(X_test_encoded)

# Evaluate the model on the 2017 test set
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Calculate FPR and FNR
tn, fp, fn, tp = conf_matrix.ravel()
fpr = fp / (fp + tn)
fnr = fn / (fn + tp)

# Print the evaluation results for the 2017 test set
print(f"Accuracy on df2017: {accuracy:.4f}")
print(f"F1 Score on df2017: {f1:.4f}")
print(f"FPR on df2017: {fpr:.4f}")
print(f"FNR on df2017: {fnr:.4f}")
print("Confusion Matrix on df2017:")
print(conf_matrix)
print("Classification Report on df2017:")
print(class_report)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.callbacks import EarlyStopping

# Assuming create_autoencoder is defined
def create_autoencoder(input_dim):
    # Define your autoencoder and encoder model here
    input_layer = tf.keras.layers.Input(shape=(input_dim,))
    encoded = tf.keras.layers.Dense(128, activation='relu')(input_layer)
    encoded = tf.keras.layers.Dense(64, activation='relu')(encoded)
    encoded = tf.keras.layers.Dense(32, activation='relu')(encoded)
    decoded = tf.keras.layers.Dense(64, activation='relu')(encoded)
    decoded = tf.keras.layers.Dense(128, activation='relu')(decoded)
    decoded = tf.keras.layers.Dense(input_dim, activation='sigmoid')(decoded)

    autoencoder = tf.keras.models.Model(inputs=input_layer, outputs=decoded)
    encoder = tf.keras.models.Model(inputs=input_layer, outputs=encoded)

    return autoencoder, encoder

def contrastive_loss(y_true, y_pred, margin=1.0):
    y_true = tf.cast(y_true, tf.float32)
    square_pred = tf.square(y_pred)
    margin_square = tf.square(tf.maximum(margin - y_pred, 0))
    return tf.reduce_mean(y_true * square_pred + (1 - y_true) * margin_square)

def generate_contrastive_pairs(X, y):
    pairs = []
    labels = []
    n = len(y)

    for i in range(n):
        # Find all positive and negative samples
        positive_indices = np.where(y == y.iloc[i])[0]
        negative_indices = np.where(y != y.iloc[i])[0]

        # Remove self-index from positive indices
        positive_indices = positive_indices[positive_indices != i]

        # Randomly select one positive and one negative sample
        if len(positive_indices) > 0:
            pos_idx = np.random.choice(positive_indices)
            pairs.append([X.iloc[i].values, X.iloc[pos_idx].values])
            labels.append(1)

        if len(negative_indices) > 0:
            neg_idx = np.random.choice(negative_indices)
            pairs.append([X.iloc[i].values, X.iloc[neg_idx].values])
            labels.append(0)

    pairs = np.array(pairs)
    labels = np.array(labels)

    return [pairs[:, 0], pairs[:, 1]], labels

# Preprocess the 2017 data
X_2017, y_2017 = preprocess_data(df2017)

# Split the 2017 data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)

# Standardize the feature data (important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert X_train and y_train to DataFrame and Series for proper indexing
X_train_df = pd.DataFrame(X_train, columns=X_2017.columns)
y_train_series = pd.Series(y_train)

# Generate contrastive pairs for training
[pairs_a, pairs_b], labels = generate_contrastive_pairs(X_train_df, y_train_series)

# Define the autoencoder
input_dim = X_train.shape[1]
autoencoder, encoder = create_autoencoder(input_dim)
autoencoder.compile(optimizer='adam', loss=MeanSquaredError())

# Define a custom training loop for contrastive learning
class ContrastiveAutoencoder(tf.keras.Model):
    def __init__(self, encoder, autoencoder, margin=1.0):
        super(ContrastiveAutoencoder, self).__init__()
        self.encoder = encoder
        self.autoencoder = autoencoder
        self.margin = margin

    def compile(self, optimizer, loss_fn):
        super(ContrastiveAutoencoder, self).compile()
        self.optimizer = optimizer
        self.loss_fn = loss_fn

    def call(self, inputs, training=False):
        input_a, input_b = inputs
        encoded_a = self.encoder(input_a)
        encoded_b = self.encoder(input_b)
        reconstructed_a = self.autoencoder(input_a)
        reconstructed_b = self.autoencoder(input_b)
        return encoded_a, encoded_b, reconstructed_a, reconstructed_b

    def train_step(self, data):
        input_a, input_b = data[0]
        y = data[1]

        with tf.GradientTape() as tape:
            # Forward pass
            encoded_a = self.encoder(input_a)
            encoded_b = self.encoder(input_b)
            reconstructed_a = self.autoencoder(input_a)
            reconstructed_b = self.autoencoder(input_b)

            # Compute reconstruction loss
            recon_loss = self.loss_fn(input_a, reconstructed_a) + self.loss_fn(input_b, reconstructed_b)

            # Compute contrastive loss
            dist = tf.reduce_sum(tf.square(encoded_a - encoded_b), axis=1)
            contrastive_loss_value = contrastive_loss(y, dist, self.margin)

            # Total loss
            total_loss = recon_loss + contrastive_loss_value

        # Compute gradients
        gradients = tape.gradient(total_loss, self.autoencoder.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.autoencoder.trainable_variables))

        return {"loss": total_loss, "recon_loss": recon_loss, "contrastive_loss": contrastive_loss_value}

# Initialize the contrastive autoencoder
contrastive_autoencoder = ContrastiveAutoencoder(encoder, autoencoder)
contrastive_autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss_fn=MeanSquaredError())

# Train the model
contrastive_autoencoder.fit(
    x=[pairs_a, pairs_b],
    y=labels,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3)]
)

# Extract embeddings for training and testing data
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# Initialize the MLP model
mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=300, random_state=42)

# Train the MLP model on the encoded training data
mlp.fit(X_train_encoded, y_train)

# Make predictions on the encoded test data
y_pred = mlp.predict(X_test_encoded)

# Evaluate the model on the 2017 test set
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Calculate FPR and FNR
tn, fp, fn, tp = conf_matrix.ravel()
fpr = fp / (fp + tn)
fnr = fn / (fn + tp)

# Print the evaluation results for the 2017 test set
print(f"Accuracy on df2017: {accuracy:.4f}")
print(f"F1 Score on df2017: {f1:.4f}")
print(f"FPR on df2017: {fpr:.4f}")
print(f"FNR on df2017: {fnr:.4f}")
print("Confusion Matrix on df2017:")
print(conf_matrix)
print("Classification Report on df2017:")
print(class_report)

# Preprocess the 2018 data
X_2018, y_2018 = preprocess_data(df2018)

# Standardize the 2018 data
X_2018 = scaler.transform(X_2018)

# Active Learning on 2018 Data
# Select an initial random subset for evaluation (e.g., 5% of the data)
initial_sample_size = int(0.05 * len(X_2018))
initial_sample_indices = np.random.choice(len(X_2018), size=initial_sample_size, replace=False)
X_2018_initial_sample = X_2018[initial_sample_indices]
y_2018_initial_sample = y_2018.iloc[initial_sample_indices]
# Remove the initial sample from the 2018 data
X_2018_remaining = np.delete(X_2018, initial_sample_indices, axis=0)
y_2018_remaining = y_2018.iloc[np.setdiff1d(np.arange(len(y_2018)), initial_sample_indices)]

# Encode the initial 2018 sample
X_2018_initial_sample_encoded = encoder.predict(X_2018_initial_sample)

# Make predictions and calculate uncertainty (e.g., entropy)
probs = mlp.predict_proba(X_2018_initial_sample_encoded)
uncertainty = -np.sum(probs * np.log(probs + 1e-6), axis=1)

# Select the most uncertain samples (e.g., top 1% most uncertain)
uncertainty_threshold = np.percentile(uncertainty, 99)
uncertain_indices = np.where(uncertainty >= uncertainty_threshold)[0]

# Label these uncertain samples (use ground truth from 2018 data)
X_uncertain = X_2018_initial_sample[uncertain_indices]
y_uncertain = y_2018_initial_sample.iloc[uncertain_indices]

# Combine these samples with the training data
X_train_combined = np.vstack([X_train, X_uncertain])
y_train_combined = np.concatenate([y_train, y_uncertain])

# Retrain the encoder and MLP with the new combined data
X_train_combined_df = pd.DataFrame(X_train_combined, columns=X_2017.columns)
y_train_combined_series = pd.Series(y_train_combined)

[pairs_a_combined, pairs_b_combined], labels_combined = generate_contrastive_pairs(X_train_combined_df, y_train_combined_series)

contrastive_autoencoder.fit(
    x=[pairs_a_combined, pairs_b_combined],
    y=labels_combined,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3)]
)

# Extract embeddings for the combined training data
X_train_combined_encoded = encoder.predict(X_train_combined)

# Retrain the MLP model
mlp.fit(X_train_combined_encoded, y_train_combined)

# Make predictions on the remaining 2018 data
X_2018_remaining_encoded = encoder.predict(X_2018_remaining)
y_pred_2018 = mlp.predict(X_2018_remaining_encoded)

# Evaluate the model on the remaining 2018 data
accuracy_2018 = accuracy_score(y_2018_remaining, y_pred_2018)
f1_2018 = f1_score(y_2018_remaining, y_pred_2018)
conf_matrix_2018 = confusion_matrix(y_2018_remaining, y_pred_2018)
class_report_2018 = classification_report(y_2018_remaining, y_pred_2018)

# Calculate FPR and FNR
tn_2018, fp_2018, fn_2018, tp_2018 = conf_matrix_2018.ravel()
fpr_2018 = fp_2018 / (fp_2018 + tn_2018)
fnr_2018 = fn_2018 / (fn_2018 + tp_2018)

# Print the evaluation results for the 2018 data
print(f"Accuracy on df2018: {accuracy_2018:.4f}")
print(f"F1 Score on df2018: {f1_2018:.4f}")
print(f"FPR on df2018: {fpr_2018:.4f}")
print(f"FNR on df2018: {fnr_2018:.4f}")
print("Confusion Matrix on df2018:")
print(conf_matrix_2018)
print("Classification Report on df2018:")
print(class_report_2018)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.callbacks import EarlyStopping

# Assuming create_autoencoder is defined
def create_autoencoder(input_dim):
    # Define your autoencoder and encoder model here
    input_layer = tf.keras.layers.Input(shape=(input_dim,))
    encoded = tf.keras.layers.Dense(128, activation='relu')(input_layer)
    encoded = tf.keras.layers.Dense(64, activation='relu')(encoded)
    encoded = tf.keras.layers.Dense(32, activation='relu')(encoded)
    decoded = tf.keras.layers.Dense(64, activation='relu')(encoded)
    decoded = tf.keras.layers.Dense(128, activation='relu')(decoded)
    decoded = tf.keras.layers.Dense(input_dim, activation='sigmoid')(decoded)

    autoencoder = tf.keras.models.Model(inputs=input_layer, outputs=decoded)
    encoder = tf.keras.models.Model(inputs=input_layer, outputs=encoded)

    return autoencoder, encoder

def contrastive_loss(y_true, y_pred, margin=1.0):
    y_true = tf.cast(y_true, tf.float32)
    square_pred = tf.square(y_pred)
    margin_square = tf.square(tf.maximum(margin - y_pred, 0))
    return tf.reduce_mean(y_true * square_pred + (1 - y_true) * margin_square)

def generate_contrastive_pairs(X, y):
    pairs = []
    labels = []
    n = len(y)

    for i in range(n):
        # Find all positive and negative samples
        positive_indices = np.where(y == y.iloc[i])[0]
        negative_indices = np.where(y != y.iloc[i])[0]

        # Remove self-index from positive indices
        positive_indices = positive_indices[positive_indices != i]

        # Randomly select one positive and one negative sample
        if len(positive_indices) > 0:
            pos_idx = np.random.choice(positive_indices)
            pairs.append([X.iloc[i].values, X.iloc[pos_idx].values])
            labels.append(1)

        if len(negative_indices) > 0:
            neg_idx = np.random.choice(negative_indices)
            pairs.append([X.iloc[i].values, X.iloc[neg_idx].values])
            labels.append(0)

    pairs = np.array(pairs)
    labels = np.array(labels)

    return [pairs[:, 0], pairs[:, 1]], labels

# Preprocess the 2017 data
X_2017, y_2017 = preprocess_data(df2017)

# Split the 2017 data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_2017, y_2017, test_size=0.2, random_state=42)

# Standardize the feature data (important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert X_train and y_train to DataFrame and Series for proper indexing
X_train_df = pd.DataFrame(X_train, columns=X_2017.columns)
y_train_series = pd.Series(y_train)

# Generate contrastive pairs for training
[pairs_a, pairs_b], labels = generate_contrastive_pairs(X_train_df, y_train_series)

# Define the autoencoder
input_dim = X_train.shape[1]
autoencoder, encoder = create_autoencoder(input_dim)
autoencoder.compile(optimizer='adam', loss=MeanSquaredError())

# Define a custom training loop for contrastive learning
class ContrastiveAutoencoder(tf.keras.Model):
    def __init__(self, encoder, autoencoder, margin=1.0):
        super(ContrastiveAutoencoder, self).__init__()
        self.encoder = encoder
        self.autoencoder = autoencoder
        self.margin = margin

    def compile(self, optimizer, loss_fn):
        super(ContrastiveAutoencoder, self).compile()
        self.optimizer = optimizer
        self.loss_fn = loss_fn

    def call(self, inputs, training=False):
        input_a, input_b = inputs
        encoded_a = self.encoder(input_a)
        encoded_b = self.encoder(input_b)
        reconstructed_a = self.autoencoder(input_a)
        reconstructed_b = self.autoencoder(input_b)
        return encoded_a, encoded_b, reconstructed_a, reconstructed_b

    def train_step(self, data):
        input_a, input_b = data[0]
        y = data[1]

        with tf.GradientTape() as tape:
            # Forward pass
            encoded_a = self.encoder(input_a)
            encoded_b = self.encoder(input_b)
            reconstructed_a = self.autoencoder(input_a)
            reconstructed_b = self.autoencoder(input_b)

            # Compute reconstruction loss
            recon_loss = self.loss_fn(input_a, reconstructed_a) + self.loss_fn(input_b, reconstructed_b)

            # Compute contrastive loss
            dist = tf.reduce_sum(tf.square(encoded_a - encoded_b), axis=1)
            contrastive_loss_value = contrastive_loss(y, dist, self.margin)

            # Total loss
            total_loss = recon_loss + contrastive_loss_value

        # Compute gradients
        gradients = tape.gradient(total_loss, self.autoencoder.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.autoencoder.trainable_variables))

        return {"loss": total_loss, "recon_loss": recon_loss, "contrastive_loss": contrastive_loss_value}

# Initialize the contrastive autoencoder
contrastive_autoencoder = ContrastiveAutoencoder(encoder, autoencoder)
contrastive_autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss_fn=MeanSquaredError())

# Train the model
contrastive_autoencoder.fit(
    x=[pairs_a, pairs_b],
    y=labels,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3)]
)

# Extract embeddings for training and testing data
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# Initialize the MLP model
mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=300, random_state=42)

# Train the MLP model on the encoded training data
mlp.fit(X_train_encoded, y_train)

# Make predictions on the encoded test data
y_pred = mlp.predict(X_test_encoded)

# Evaluate the model on the 2017 test set
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Calculate FPR and FNR
tn, fp, fn, tp = conf_matrix.ravel()
fpr = fp / (fp + tn)
fnr = fn / (fn + tp)

# Print the evaluation results for the 2017 test set
print(f"Accuracy on df2017: {accuracy:.4f}")
print(f"F1 Score on df2017: {f1:.4f}")
print(f"FPR on df2017: {fpr:.4f}")
print(f"FNR on df2017: {fnr:.4f}")
print("Confusion Matrix on df2017:")
print(conf_matrix)
print("Classification Report on df2017:")
print(class_report)

# Preprocess the 2018 data
X_2018, y_2018 = preprocess_data(df2018)

# Standardize the 2018 data
X_2018 = scaler.transform(X_2018)

# Active Learning on 2018 Data
# Select a random 1% sample of the 2018 data
sample_size = int(0.01 * len(X_2018))
random_indices = np.random.choice(len(X_2018), size=sample_size, replace=False)
X_2018_sample = X_2018[random_indices]
y_2018_sample = y_2018.iloc[random_indices]

# Remove the sampled data from the 2018 data
X_2018_remaining = np.delete(X_2018, random_indices, axis=0)
y_2018_remaining = y_2018.iloc[np.setdiff1d(np.arange(len(y_2018)), random_indices)]

# Encode the 2018 sample
X_2018_sample_encoded = encoder.predict(X_2018_sample)

# Combine the 2018 sample with the training data
X_train_combined = np.vstack([X_train_encoded, X_2018_sample_encoded])
y_train_combined = np.concatenate([y_train, y_2018_sample])

# Retrain the MLP model
mlp.fit(X_train_combined, y_train_combined)

# Make predictions on the remaining 2018 data
X_2018_remaining_encoded = encoder.predict(X_2018_remaining)
y_pred_2018 = mlp.predict(X_2018_remaining_encoded)

# Evaluate the model on the remaining 2018 data
accuracy_2018 = accuracy_score(y_2018_remaining, y_pred_2018)
f1_2018 = f1_score(y_2018_remaining, y_pred_2018)
conf_matrix_2018 = confusion_matrix(y_2018_remaining, y_pred_2018)
class_report_2018 = classification_report(y_2018_remaining, y_pred_2018)

# Calculate FPR and FNR
tn_2018, fp_2018, fn_2018, tp_2018 = conf_matrix_2018.ravel()
fpr_2018 = fp_2018 / (fp_2018 + tn_2018)
fnr_2018 = fn_2018 / (fn_2018 + tp_2018)

# Print the evaluation results for the 2018 data
print(f"Accuracy on df2018: {accuracy_2018:.4f}")
print(f"F1 Score on df2018: {f1_2018:.4f}")
print(f"FPR on df2018: {fpr_2018:.4f}")
print(f"FNR on df2018: {fnr_2018:.4f}")
print("Confusion Matrix on df2018:")
print(conf_matrix_2018)
print("Classification Report on df2018:")
print(class_report_2018)
```
